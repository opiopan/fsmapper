```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# draw a graph
plt.style.use('dark_background')
plt.rcParams["font.size"] = 18
plt.rcParams["figure.facecolor"] = '#777EAB'
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot()
ax.xaxis.set_major_locator(mpl.ticker.MultipleLocator(0.5))
ax.yaxis.set_major_locator(mpl.ticker.MultipleLocator(0.5))
ax.grid(True, color='#606060', linewidth=3)
ax.axhline(0, color='gray', linewidth=3)
ax.axvline(0, color='gray', linewidth=3)

# Parameters
d = 0.0 # width of deadzone
a = 2.5 # growth
# input range
x = np.linspace(-1, 1, 400)

a = 0.00001
max_val = np.exp(a * (1 - d)) - 1
y = np.where(np.abs(x) <= d, 0, np.sign(x) * (np.exp(a * (np.abs(x) - d)) - 1) / max_val)
plt.plot(x, y, label=f'Growth Rate = {a}', linewidth=3, linestyle='dashed')

a = 2
max_val = np.exp(a * (1 - d)) - 1
y = np.where(np.abs(x) <= d, 0, np.sign(x) * (np.exp(a * (np.abs(x) - d)) - 1) / max_val)
plt.plot(x, y, label=f'Growth Rate = {a}', linewidth=6)

a = 5
max_val = np.exp(a * (1 - d)) - 1
y = np.where(np.abs(x) <= d, 0, np.sign(x) * (np.exp(a * (np.abs(x) - d)) - 1) / max_val)
plt.plot(x, y, label=f'Growth Rate = {a}', linewidth=6)

plt.show()
```
